```python
import numpy as np
import matplotlib.pyplot as plt
import methods
```

```python
n = 200
dim = 1
np.random.seed(12)
X = 3 * np.random.rand(n, dim)
y = 6 * X + 1 * np.random.randn(n, dim)

learning_rate = methods.const_learning_rate
batch_sizes = [1, 50, 70, 120, 200]

print("SGD:")
colors = ['r', 'orange', 'y', 'g', 'violet']
for i, batch_size in enumerate(batch_sizes):
    w, iters = methods.stochastic_gradient_descent(X, y, learning_rate=learning_rate, batch_size=batch_size, eps=1e-2, max_iter=10000)
    print("batch: {}. w: {}. iters: {}".format(batch_size, w, iters))
    plt.plot(X, y, 'b.')
    plt.plot(X, X.dot(w), '-', color=colors[i], label=f'Batch size {batch_size}', linewidth=3)
    plt.legend()
    plt.title('sgd')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```
