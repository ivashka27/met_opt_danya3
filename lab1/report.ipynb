```python
import numpy as np
import method.gradient_descent as gd
import method.dichotomy_gradient as dg
import method.wolfe_gradient as wg
import plot.plotter as plotter
```

```python
def f(x):
    #return 10 * x[0] ** 2 + x[1] ** 2
    return 4 * x[0] ** 2 + 9 * x[1] ** 2 - 4 * x[0] * x[1] - 2 * x[0] + 12 * x[1] + 7


def grad(x):
    h = 1e-5
    return (f(x[:, np.newaxis] + h * np.eye(2)) - f(x[:, np.newaxis] - h * np.eye(2))) / (2 * h)
```

# $f(x, y) = 10x^2 + y^2$
# $f(x, y) = 4x^2 + 9y^2 - 4xy - 2x + 12y + 7$

```python
def plot_from_start(start):
    lr = 0.05
    (points1, grad_calc1, func_calc1) = gd.gradient_descent(f, grad, start, 1e-4, lr, 100000)
    print("start:", start)
    print("gradient descent:")
    print(points1[-1], f(points1[-1]))
    print("gradient calculations:", grad_calc1)
    print("function calculations:", func_calc1)

    (points2, grad_calc2, func_calc2) = dg.dichotomy_gradient(f, grad, start, 1e-4, 1e-3, 0.5)
    print("dichotomy gradient:")
    print(points2[-1], f(points2[-1]))
    print("gradient calculations:", grad_calc2)
    print("function calculations:", func_calc2)

    (points3, grad_calc3, func_calc3) = wg.wolfe_gradient(f, grad, start, 1e-4)
    print("wolfe gradient:")
    print(points3[-1], f(points3[-1]))
    print("gradient calculations:", grad_calc3)
    print("function calculations:", func_calc3)

    plotter.multiple_points_over_contour(points1, points2, points3, f, lr).show()
```

```python
plot_from_start([-1, 2])
```

```python
plot_from_start([-0.5, -1])
```

```python
plot_from_start([-0.5, 0])
```
