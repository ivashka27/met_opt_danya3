```python
import matplotlib.pyplot as plt
import numpy as np
from random import uniform
import method.gradient_descent as gd
import method.dichotomy_gradient as dg
import plot.plotter as plotter
```

```python
def f(x):
    return 10 * x[0] ** 2 + x[1] ** 2


def grad(x):
    h = 1e-5
    return (f(x[:, np.newaxis] + h * np.eye(2)) - f(x[:, np.newaxis] - h * np.eye(2))) / (2 * h)
```

# $f(x, y) = 10x^2 + y^2$

```python
def plot_from_start(start):
    (points1, grad_calc1, func_calc1) = gd.gradient_descent(f, grad, start, 1e-6, 0.06, 100000)
    print("gradient descent:")
    print(points1[-1], f(points1[-1]))
    print("gradient calculations:", grad_calc1)
    print("function calculations:", func_calc1)

    (points2, grad_calc2, func_calc2) = dg.dichotomy_gradient(f, grad, start, 1e-6, 1e-3, 0.5)
    print("dichotomy gradient:")
    print(points2[-1], f(points2[-1]))
    print("gradient calculations:", grad_calc2)
    print("function calculations:", func_calc2)

    plotter.multiple_points_over_contour(points1, points2, f, 0.06).show()
```

```python
plot_from_start([-1, 2])
```

```python
plot_from_start([-0.5, -1])
```
