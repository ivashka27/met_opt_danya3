```python
import matplotlib.pyplot as plt
import numpy as np
import lab1.method.gradient_descent as gd
import lab1.method.dichotomy_gradient as dg
import lab1.plot.plotter as plotter
```

```python
def f(x):
    return 10 * x[0] ** 2 + x[1] ** 2


def grad(x):
    h = 1e-5
    return (f(x[:, np.newaxis] + h * np.eye(2)) - f(x[:, np.newaxis] - h * np.eye(2))) / (2 * h)
```

```python
start = [1, 2]

(points1, grad_calc1, func_calc1) = gd.gradient_descent(f, grad, start, 1e-6, 0.04, 100000)
pl1 = plotter.points_over_contour(points1, f)
pl1.title("gradient descent")
pl1.show()
print(points1[-1], f(points1[-1]))

(points2, grad_calc2, func_calc2)  = dg.dichotomy_gradient(f, grad, start, 1e-6, 1e-3, 1)
pl2 = plotter.points_over_contour(points2, f)
pl2.title("dichotomy descent")
pl2.show()

print()
```
